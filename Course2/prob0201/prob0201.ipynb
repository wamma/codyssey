### 1. 프로젝트 목표/요구사항 정리하기
- 현재 문제의 **최종 목표**: MovieLens, KMRD, Netflix 공개데이터 중에서 내가 개발하고자 하는 추천 시스템에 맞는 데이터를 고름
- 이를 위해서 각 데이터셋이 어떤 필드를 제공하는지와 추천 모델에 필요한 정보가 잘 들어 있는지를 확인해야 함

> 즉, 이 단계에서는 "추천 모델을 만들려면 어떤 정보가 꼭 필요할까?" 를 먼저 리스트업

### 2. 각 데이터셋(MovieLens, KMRD, Netflix)에 대한 조사
이 단계에서는 각 공개 데이터셋에 대해 다음과 같은 정보를 찾고/정리한다.
1. 데이터셋 개요
	- 이름: MovieLens, KMRD, Netflix
	- 데이터 출처, 규모(예: 사용자 수, 영화 수, 평점 수)
2. 제공하는 주요 필드(컬럼)
	- 예: movieId, title, genres, rating, timestamp, actors(배우), country(제작국가), releaseYear, ratingDate 등등
	- 해당 컬럼이 내가 만든 추천 시스템에 얼마나 도움이 되는지(Useful/Not Useful) 등을 체크
3. 라이선스/사용제한
	- 실제 서비스나 연구목적으로 사용할 때 저작권/사용권 문제가 없는지?
4. 데이터 형태(포맷)
	- CSV, TSV, JSON 등인지, 테이블 구조는 어떻게 되어 있는지
	- 전처리 난이도가 어느 정도인지

### 3. 각 데이터셋 특징 비교
##### 1. 규모
- MovieLens
	- 버전에 따라 100K부터 32M 이상의 평점 데이터를 제공
	- 연구용/실습용으로 100K, 1M, 20M, 25M 등이 가장 유명하고 많이 쓰임

- KMRD
	- small: 약 13만 건의 평점
	- 2M: 약 250만 건 평점
	- 5M: 약 500만 건 평점
	- 한국 사용자/영화 중심으로, 비교적 규모가 중간 정도

- Netflix
	- 'Netflix Prize'용 데이터: 1억 건 이상 평점 (사용자 약 48만명, 영화 1만 8천여 편)
	- 매우 대규모, 스케일업 테스트나 딥러닝용 대용량 학습에 유리
> 스케일업 테스트는 ‘어떤 시스템(또는 모델)이 더 큰 규모에서도 안정적이고 빠르게 동작할 수 있는지’ 확인

##### 2. 메타데이터 제공 여부
- MovieLens
	- 영화 제목, 장르, 날짜, 태그 정도
	- 베우/감독/제작국가 정보는 포함되지 않음 -> 별도 IMDb, TMDB, API 연동 필요
	- Tag Genome 데이터셋 등을 추가로 제공하기도 함

- KMRD
	- 한국 영화 중심 메타데이터를 상세하게 제공 (배우, 장르, 국가, 연도, 등급 등)
	- 사용자가 같은 영화를 중복 평가한 경우가 존재하므로, 전처리에 주의가 필요
	- Naver Movie Scraper로부터 수집 -> 한국 영화 분석 프로젝트에 유리

- Netflix
	- 평점 정보가 주로 공개 (사용자 Id, 영화 Id, 평점, 시점)
	- 영화별 구체 메타데이터(배우, 감독, 장르)는 별도 소스 매핑 필요
	- Netflix 자체 이름이지만, 실제 공개된 데이터는 철저히 익명화된 평점 정보가 전부

### 3. 대표적 활용사례
- MovieLens
	- 가장 흔히 보는 추천시스템 예제. 협업 필터링, 행렬분해, 딥러닝 등 학계/튜토리얼 표준
	- 규모가 다 달라 스케일 업 테스트 가능

- KMRD
	- 한국 영화 중심으로 배우, 국가 등 다양한 메타데이터 추천을 테스트할 때 적합
	- 한국영화만 보는데, 배구 기반 협업필터링을 해보고 싶다 같은 니즈에 유리

- Netflix
	- 대규모 협업필터링, 경진대회(Netflix Prize)등에서 활용
	- 실제 대용량 환경에서 분산 처리(Spark 등) 실습에 적합
