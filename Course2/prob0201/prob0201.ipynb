### 영화 관련 공개 데이터셋에 대해 조사한다 (목표)
각 데이터셋의 구성 요소와 제공 정보 확인

#### 1. MovieLens
##### **장점**
- 다양한 크기의 데이터셋 제공 (100K부터 32M까지)
- 풍부한 메타데이터
	- 영화 정보 (제목, 장르, 출시일)
	- 사용자 평점
	- 태그 시스템
	- 태그 게놈 데이터 (영화-태크 관련성 점수)
	- 정기적인 업데이트
	- 안정적인 벤치마크 데이터셋
	- 무료로 사용 가능
##### **제공 정보**
- 영화, ID, 제목, 장르
- 사용자 ID, 평점, 평가 시간
- 태그 데이터
- IMDB, TMDB 링크


2. KMRD: 한국 영화 데이터에 초점(평점, 배우, 장르 등)

3. Netflix: 과거 공개된 데이터, 사용자와 영화 매핑(현재는 구하기 어려움)


##### 각 데이터에서 제공하는 항목이 개발하려고 하는 추천 시스템에서 활용하기 쉬운지 파악
#### 영화, 사용자, 장르, 출연배우, 제작국가, 제작일, 출시일, 평점, 평가일 등 제공하는 정보를 확인하고 조합해 보면서 확인이 필요하다.
