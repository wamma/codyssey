사용자별로 몇 편의 영화에 평점을 부여했는지 분포를 확인하라. (Histogram)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

rates = pd.read_csv("kmrd-small/rates.csv")
```

```python
user_movie_count = rates.groupby('user').size()

plt.figure(figsize=(10, 6))
plt.hist(user_movie_count, bins=10)
plt.title('Number of Movies Rated by Users')
plt.xlabel('Number of Movies Rated')
plt.ylabel('Number of Users')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
# 초 단위로 변환
rates['time'] = pd.to_datetime(rates['time'], unit='s')

monthly_ratings = rates.resample('ME', on='time').size()


plt.figure(figsize=(10, 6))
plt.plot(monthly_ratings.index, monthly_ratings.values)
plt.title('Monthly Ratings')
plt.xlabel('Month')
plt.ylabel('Number of Ratings')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
